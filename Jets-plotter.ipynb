```python
import torch
import matplotlib.pyplot as plt 
from scipy.stats import norm
from matplotlib.offsetbox import AnchoredText
from matplotlib.gridspec import GridSpec

#GPU
device = 'cpu'
print(f"Using {device} device")
jets, targets = torch.load('/data/gabrieleb/data/Jets/dataset_test', map_location = device)
```

```python
fig, ax2 = plt.subplots(1,4,figsize = (28,7))

title1 = '1 prong'
title2 = '2 prongs'
title3 = '3 prongs'
title4 = '4 prongs'

xtitle = '\u03C6'
ytitle = '\u03B7'

fig.suptitle('Realistic jets', fontsize = 40)

ax2[0].set_title(title1, fontsize = 20)
ax2[0].set_xlabel(xtitle, fontsize = 20)
ax2[0].set_ylabel(ytitle, fontsize = 20)
ax2[0].tick_params(axis='both', which='major', labelbottom = False, labelleft = False)
# ax2[0].set_xlim(-0.1, 0)
# ax2[0].set_ylim(-0.5, 1)

ax2[1].set_title(title2, fontsize = 20)
ax2[1].set_xlabel(xtitle, fontsize = 20)
ax2[1].set_ylabel(ytitle, fontsize = 20)
ax2[1].tick_params(axis='both', which='major', labelbottom = False, labelleft = False)
# ax2[1].set_xlim(-0.50, -0.1)
# ax2[1].set_ylim(-0.4, 3)

ax2[2].set_title(title3, fontsize = 20)
ax2[2].set_xlabel(xtitle, fontsize = 20)
ax2[2].set_ylabel(ytitle, fontsize = 20)
ax2[2].tick_params(axis='both', which='major', labelbottom = False, labelleft = False)
# ax2[2].set_xlim(-0.6, -0.3)
# ax2[2].set_ylim(-0.2, 0.3)

ax2[3].set_title(title4, fontsize = 20)
ax2[3].set_xlabel(xtitle, fontsize = 20)
ax2[3].set_ylabel(ytitle, fontsize = 20)
ax2[3].tick_params(axis='both', which='major', labelbottom = False, labelleft = False)
# ax2[3].set_xlim(-0.6, -0.35)
# ax2[3].set_ylim(-0.10, 0.2)

for i in range(jets.size(1)):
    ax2[0].scatter(jets[torch.nonzero(targets == 1)[:1].item()][i][1], jets[torch.nonzero(targets == 1)[:1].item()][i][2], s = 10*jets[torch.nonzero(targets == 1)[:1].item()][i][0], color = 'b', alpha = 0.3)
    ax2[1].scatter(jets[torch.nonzero(targets == 2)[:1].item()][i][1], jets[torch.nonzero(targets == 2)[:1].item()][i][2], s = 10*jets[torch.nonzero(targets == 2)[:1].item()][i][0], color = 'b', alpha = 0.3)
    ax2[2].scatter(jets[torch.nonzero(targets == 3)[:1].item()][i][1], jets[torch.nonzero(targets == 3)[:1].item()][i][2], s = 10*jets[torch.nonzero(targets == 3)[:1].item()][i][0], color = 'b', alpha = 0.3)
    ax2[3].scatter(jets[torch.nonzero(targets == 4)[:1].item()][i][1], jets[torch.nonzero(targets == 4)[:1].item()][i][2], s = 10*jets[torch.nonzero(targets == 4)[:1].item()][i][0], color = 'b', alpha = 0.3)
    
fig.savefig('/home/gabrieleb/Plot/Jets/Real/jets_real.pdf', facecolor='w',pad_inches=0.3, transparent=True)
```
